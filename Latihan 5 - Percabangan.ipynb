```python
# Break Statement
x = ['saya', 'suka', 'rpl', 'tapi', 'boong']
for word in x:
    if word == 'tapi':
       break
    # lengkapi kode di bawah ini
    
    print(word,end='')
    
# buatkan output menjadi saya suka rpl
```
